# EXPLORING QUALITY OF LIFE IN CHILDREN WITH AUTISM: A COMPARATIVE ANALYSIS WITH TYPICALLY DEVELOPING CHILDREN

**Objective**: To explore differences in quality of life between children with ASD and typically developing (TD) children, and to analyze demographic factors (such as gender, parental education, and family structure) influencing QoL outcomes within the ASD group.

### Dataset Overview:
The dataset consists of 150 rows and 22 columns, that covers a wide range of variables, including demographic, psychosocial and physiological factors. It includes the following details; Group, Age,	Gender,	Socioeconomic Status,	Parental Education,	Severity of ASD Symptoms,	Comorbid Conditions,	Family Structure,	Parental Mental Health Issues,	Type of Schooling, Therapy Received, Daily Living Skills, Social Skills,	Communication Skills,	School Support Services,	Peer Relationships,	Social Inclusion,	Physical Health,	Emotional Well-being,	Social Functioning,	School Functioning,	Overall Quality of Life

### Key Features of The Dataset:
+ Quality of Life Metrics: Physical Health, Emotional Well-being, Social Functioning, School Functioning, Overall Quality of Life
+ Demographics: Age, Gender, Socioeconomic Status, Parental Education, Family Structure
+ Intervention: Type of Schooling, Therapy Received, School Support Services
+ Psychosocial Factors: Daily Living Skills, Social Skills, Communication Skills, Peer Relationships, Social Inclusion
+ Physiological Factors: Severity of ASD Symptoms, Comorbid Conditions, Parental Mental Health Issues

Summary of the value counts and percentage of each categorical variable

```python
def combined_categorical_summary(df, group_col='Group', output_file='categorical_statistical_summary.xlsx'):
    group_labels = df[group_col].dropna().unique()
    categorical_cols = df.select_dtypes(include='object').columns
    categorical_cols = [col for col in categorical_cols if col != group_col]

    writer = pd.ExcelWriter(output_file, engine='xlsxwriter')

    for col in categorical_cols:
        table = pd.DataFrame()

        for group in group_labels:
            group_df = df[df[group_col] == group]
            counts = group_df[col].value_counts(dropna=False)
            percents = group_df[col].value_counts(normalize=True, dropna=False) * 100

            group_table = pd.DataFrame({
                f'{group} Count': counts,
                f'{group} %': percents.round(2)
            })

            table = table.join(group_table, how='outer')

        table = table.fillna(0).astype({f'{g} Count': 'int' for g in group_labels})
        table.to_excel(writer, sheet_name=col[:31])  # Excel sheet name limit = 31 chars

    writer.close()
    print(f"Combined summary saved to '{output_file}'")

    return table

# Call the function
combined_categorical_summary(df)
```

Statistical test to compare the P value and mean across the two groups, ASD and TD on the numerical variables

```python
def compare_continuous(df, group_col='Group', group1='ASD', group2='TD', alpha=0.05):
    numeric_cols = df.select_dtypes(include='number').columns
    results = []

    for col in numeric_cols:
        group1_data = df[df[group_col] == group1][col].dropna()
        group2_data = df[df[group_col] == group2][col].dropna()

        # Normality test
        p1 = shapiro(group1_data)[1]
        p2 = shapiro(group2_data)[1]
        normal = (p1 > alpha) and (p2 > alpha)

        if normal:
            test_stat, p_value = ttest_ind(group1_data, group2_data, equal_var=False)
            test_used = 'T-test'
        else:
            test_stat, p_value = mannwhitneyu(group1_data, group2_data, alternative='two-sided')
            test_used = 'Mann-Whitney U'

        results.append({
            'Variable': col,
            'P-Value': round(p_value, 4),
            'ASD_Mean': round(group1_data.mean(), 2),
            'TD_Mean': round(group2_data.mean(), 2)
        })

    result_df = pd.DataFrame(results)
    result_df.to_excel('T-test.xlsx', index=False)
    print("Continuous group comparison saved to 'continuous_group_comparison.xlsx'")
    return test_used, result_df

# Call the function
cont_comp = compare_continuous(df)
cont_comp
```

+ **Physical Health (p < 0.001):** Children with ASD reported signigicantly lower physical health scores (M = 64.97) compared to their TD peers (M = 85.56).
+ **Emotional Health (p < 0.001):** Children with ASD had significantly lower emotional health scores (M = 57.53) than TD children (M = 79.00).  
+ **Social Functioning (p < 0.001):** Children with ASD had significantly lower social functioning scores ( M = 49.11) compared to their TD peers (M = 82.26).
+ **School Functioning (p < 0.001):** Children with ASD had lower school functioning scores (M = 52.19) than TD children (M = 77.63).
+ **Overall Quality of Life (p < 0.001):** The ASD group reported significantly lower overall QoL scores (M = 56.96) than the TD group (M = 82.29).  


 Chi Square test for statistical significance on the continuous variables

```python
def compare_categorical(df, group_col='Group', alpha=0.05):
    cat_cols = df.select_dtypes(include='object').columns.drop(group_col)
    results = []

    for col in cat_cols:
        contingency = pd.crosstab(df[col], df[group_col])
        if contingency.shape[1] == 2 and contingency.shape[0] >= 2:
            chi2, p, _, _ = chi2_contingency(contingency)
            results.append({
                'Variable': col,
                'P-Value': round(p, 4),
                'Chi2 Stat': round(chi2, 2)
            })

    result_df = pd.DataFrame(results)
    return result_df

# Call the function
cat_comp = compare_categorical(df, group_col='Group')

cat_comp = cat_comp.sort_values(by='Chi2 Stat', ascending=False)

cat_comp.to_excel('Chi-Test.xlsx', index=False)
cat_comp
```

Several variables showed statistically significant differences between the ASD and TD groups (p < 0.005), indicating that these characteristics are meaningfully associated with whether a child is diagnosed with ASD.

The four variables with the highest chi-square values were:
+ Severity
+ Comorbid Conditions
+ Parent Mental Health Issues
+ Therapy received.

These variables were only applicable to children with ASD, which explains their high statistical significance.

Among the variables applicable to both groups, those with the highest statistical significance were:
+ QoL Level
+ Peer Relationships
+ Living Skills
+ Social Skills

This suggests that there are clear and meaningful differences between ASD and TD children in these areas.

On the other hand, the following variables showed no statistical significance:
+ Socioeconomic Status
+ Parental Education
+ Gender

This means there is no strong evidence of a real difference between ASD and TD children in these characteristics; any observed differences are likely due to random variation.

### Correlational Analysis

```python
def compute_eta_squared(H, n):
    """Effect size for Kruskal-Wallis (eta squared)"""
    return H / (n - 1)

def compute_rank_biserial(u_stat, n1, n2):
    """Effect size for Mann-Whitney (rank biserial)"""
    return 1 - (2 * u_stat) / (n1 * n2)

def median_qol_with_effect_size(
    data,
    group_col='Group',
    target='Overall_QoL',
    significance=0.05
):
    asd_df = data[data[group_col] == 'ASD']
    categorical_cols = [
        col for col in asd_df.select_dtypes(include='object').columns
        if col not in [group_col, target]
    ]

    results = []

    for col in categorical_cols:
        categories = asd_df[col].dropna().unique()
        if len(categories) < 2:
            continue  # skip variables with <2 categories

        groups = [asd_df[asd_df[col] == cat][target] for cat in categories]

        # Compute medians
        medians = [int(np.median(g)) for g in groups]
        medians_display = ' / '.join([str(m) for m in medians])

        # Detailed IQRs
        detailed = []
        for cat, g in zip(categories, groups):
            med = int(np.median(g))
            q1 = int(np.percentile(g, 25))
            q3 = int(np.percentile(g, 75))
            detailed.append(f"{cat}: {med} ({q1}–{q3})")

        # Statistical Test + Effect Size
        if len(categories) == 2:
            stat, p = mannwhitneyu(groups[0], groups[1], alternative='two-sided')
            n1, n2 = len(groups[0]), len(groups[1])
            effect_size = compute_rank_biserial(stat, n1, n2)
            effect_type = 'r_biserial'
        else:
            stat, p = kruskal(*groups)
            n = sum(len(g) for g in groups)
            effect_size = compute_eta_squared(stat, n)
            effect_type = 'eta²'

        results.append({
            'Variable': col.replace('_', ' '),
            'ASD Categories': ' / '.join(categories),
            'Median QoL': medians_display,
           # 'IQR Details': ' / '.join(detailed),
            'p-value': f"{p:.3f}" + ('*' if p < significance else ''),
            'Effect Size': f"{effect_size:.3f}"
        })
    return pd.DataFrame(results)

summary_df = median_qol_with_effect_size(df)
summary_df = summary_df.sort_values(by='p-value')
summary_df.to_excel('Association_Analysis.xlsx', index=False)
summary_df
```

Among children diagnosed with ASD, only QoL Level and Gender showed statistically significant associations with overall Quality of Life scores. Notably, males had higher median QoL scores than females, with a large effect size (0.481). All other variables, including severity, living skills, social skills, and school support, did not show significant associations, suggesting that individual experiences of QoL may not be strongly predicted by these factors in this sample.

While prior studies have shown significant associations between factors such as severity, communication, and comorbidities with QoL in children with ASD, this analysis did not find these associations to be statistically significant. This discrepancy may be due to sample characteristics, measurement differences, or unaccounted confounding variables.

Check why gender has a strong association with overall QoL

```python
# Create a contingency table (cross-tabulation)
contingency_table = pd.crosstab(df_ASD['Gender'], df_ASD['QoL_Level'])

# Run Chi-Square Test
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Print results
print("Chi-Square Test Results")
print(f"Chi2 Statistic: {chi2:.3f}")
print(f"Degrees of Freedom: {dof}")
print(f"P-value: {p:.4f}")

# Print Observed and Expected values
observed = contingency_table
expected = expected.astype(int)

print('\n', 'Observed Frequencies Table:')
print(observed)
print("\nExpected Frequencies Table:")
expected_df = pd.DataFrame(expected, index=contingency_table.index, columns=contingency_table.columns)
print(expected_df)
```

Within the ASD group, a chi-square test of independence revealed a statistically significant association between gender and Quality of Life (χ² = 8.036, df = 3, p = 0.045).
The data show that female children with ASD were more likely to fall into Very Poor or Poor QoL categories, with none reporting Fair or Good QoL. In contrast, male children were more evenly distributed, with some reporting Fair QoL.
These findings suggest that female children with ASD may experience lower quality of life compared to their male counterparts, even within the same diagnostic group.

Despite the overrepresentation of males in the ASD group, the association between gender and Quality of Life remained statistically significant. This suggests that the lower QoL observed among females is not simply due to sample imbalance, but reflects a real and concerning gender disparity in the lived experience of children with ASD.

Box plots on the numerical variables to detect outliers and to visualize the distribution

```python

import math

def plot_boxplots_grid(df, group_col='Group', ncols=3, nrows=2, save=False, save_prefix='boxplot_page'):
    """
    Generates box plots in a 2x3 grid layout for all numerical columns.
    If `group_col` exists, split by ASD vs TD.
    """
    numeric_cols = df.select_dtypes(include='number').columns.tolist()
    total_plots = len(numeric_cols)
    plots_per_fig = ncols * nrows
    num_pages = math.ceil(total_plots / plots_per_fig)

    for page in range(num_pages):
        fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(ncols*3, nrows*3))
        axes = axes.flatten()

        for i in range(plots_per_fig):
            idx = page * plots_per_fig + i
            if idx >= total_plots:
                axes[i].axis('off')
                continue

            col = numeric_cols[idx]
            if group_col in df.columns:
                sns.boxplot(x=group_col, y=col, data=df, ax=axes[i], palette='Set2')
                axes[i].set_title(f'{col}')
            else:
                sns.boxplot(y=df[col], ax=axes[i], color='skyblue')
                axes[i].set_title(f'Boxplot of {col}')

        plt.tight_layout()
        plt.savefig('numerical_box_plots', dpi=300, bbox_inches='tight')
        # if save:
        #     filename = f'{save_prefix}_{page+1}.png'
        #     plt.savefig(filename, dpi=300)
        #     print(f"Saved: {filename}")
        plt.show()
plot_boxplots_grid(df)
```

Run a Chi test on school support services vs group.

```python
# Create a contingency table
contingency_table = pd.crosstab(df['Group'], df['School_Support_Services'])

# Run Chi-Square Test
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Print results
print("Chi-Square Test Results")
print(f"Chi2 Statistic: {chi2:.3f}")
print(f"Degrees of Freedom: {dof}")
print(f"P-value: {p:.4f}")

# Print Observed and Expected values
observed = contingency_table
expected = expected.astype(int)

print('\n', 'Observed Frequencies Table:')
print(observed)
print("\nExpected Frequencies Table:")
expected_df = pd.DataFrame(expected, index=contingency_table.index, columns=contingency_table.columns)
print(expected_df)
```

This reveals a statistically significant association between group (ASD vs TD) and access to school support services.

```python
# Define a function to encode the categorical data
from sklearn.preprocessing import LabelEncoder

def label_encode_categoricals(df):
    df_encoded = df.copy()
    le = LabelEncoder()

    # Loop through columns that are not numeric
    for col in df_encoded.select_dtypes(exclude=['int64', 'float64']).columns:
        df_encoded[col] = le.fit_transform(df_encoded[col].astype(str))

    return df_encoded

df_ASD_new = label_encode_categoricals(df_ASD_new)
df_ASD_new
```
